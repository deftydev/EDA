```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

##display all the columns of the dataset
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.head()
```

```python
#display the shape of dataset in rows and columns form
print(dataset.shape)
```

```python
dataset.isnull()
```

# In Data analysis we will analyze to find the below stuff

1. Missing Values
2. All the numerical variables
3. Distribution of the numerical variables
4. Categorical Variables
5. Outliers
6. Relationship between independent and dependent features
7. Correlations

```python
dataset_missing=[]
for feature in dataset.columns:
    if dataset[feature].isnull().sum()>=1:
        dataset_missing.append(feature)
        
print(dataset_missing)
```

```python
len(dataset_missing)
```

# How to find the average of null values in each feature in dataset_missing. 

```python
for feature in dataset_missing:
    missing= dataset[feature].isnull().mean()*100
    print(f"The {feature} has {missing}% missing values")
    
```

```python
data=dataset.copy()

for feature in dataset_missing:
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    plt.show()
    
    
    

    
```

```python
categorical_feature=[]
numerical_feature=[]
for feature in dataset.columns:
    if dataset[feature].dtypes=='object':
        categorical_feature.append(feature)
    else:
        numerical_feature.append(feature)
        
print(categorical_feature)
print(numerical_feature)
    
```

```python
data=dataset.copy()
for feature in numerical_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

```python
year_feature=[]
for feature in dataset.columns:
    if 'Yr' in feature or 'Year' in feature:
        year_feature.append(feature)
        
print(year_feature)
```

```python
data=dataset.copy()

for feature in year_feature:
    if feature !='YrSold':
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
data=dataset.copy()
continous_feature=[]
for feature in numerical_feature:
    if feature =='Id':
        continue
    if len(data[feature].unique())>25:
        continous_feature.append(feature)
        
print(continous_feature)
```

```python
for feature in continous_feature:
    data[feature]=np.log(data[feature])
    
    plt.scatter(data[feature],data['SalePrice'])
    plt.show()
```

```python
## Outliers 
for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
